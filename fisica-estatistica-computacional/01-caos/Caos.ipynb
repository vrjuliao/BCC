```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.

```python
r = np.array([.5, 2.5, 3.1, 3.5, 3.7])
x0 = .5
n_max = 50

def pop_dynamic(x0, r, n_max):
    x = np.zeros((len(r), n_max))
    x[:,0] = np.repeat(x0, len(r))
    for i in range(1, 50):
        x[:,i] = r*x[:,i-1]*(1-x[:,i-1])
    return x

x = pop_dynamic(x0, r, n_max)
for i in range(len(r)):
    plt.plot(x[i], label=str(r[i]))    
    plt.legend(title='r', loc='lower right')
    plt.show()
```

## 2.

```python
q2_x025 = pop_dynamic(.25, r, n_max)
for i in range(len(r)):
    plt.plot(q2_x025[i], label=str(r[i]))

plt.legend(title='r', loc='lower right')
plt.title('x0 = {xx}'.format(xx=0.25))
plt.show()

q2_x05 = pop_dynamic(.5, r, n_max)
for i in range(len(r)):
    plt.plot(q2_x05[i], label=str(r[i]))

plt.legend(title='r', loc='lower right')
plt.title('x0 = {xx}'.format(xx=0.5))
plt.show()

q2_x075 = pop_dynamic(.75, r, n_max)
for i in range(len(r)):
    plt.plot(q2_x075[i], label=str(r[i]))

plt.legend(title='r', loc='lower right')
plt.title('x0 = {xx}'.format(xx=0.75))
plt.show()

print('40th value when x0 = 0.25:')
print(q2_x025[:, 39])

print('\n40th value when x0 = 0.75:')
print(q2_x075[:, 39])

print('\n40th value when x0 = 0.5:')
print(q2_x05[:, 39])

print('\n43th value when x0 = 0.5:')
print(q2_x05[:, 42])
```
